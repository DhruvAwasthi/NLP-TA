### Stemming

Words are reduced to their base form removing suffixes and prefixes which can be used for information retrieval and text analysis. For example, swimmer might get reduced to swim.  

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()

sentence = """Stemming is a technique where words are reduced to their root or base form, removing suffixes or prefixes"""
tokenized_sentence = word_tokenize(sentence)

final_sentence = [stemmer.stem(token) for token in tokenized_sentence]

print("Original tokens:\n",tokenized_sentence,"\n")
print("After Porter stemmer:\n",final_sentence)
```

```python
words = ['swim', 'swam', 'swimming', 'swimmer']

for w in words:
    print(w, " \t: ", stemmer.stem(w))
```

### Lemmatization

Breaking a word down to its root meaning, to identify similarities. Here, base word is reffered as a lemma. Very similar to stemming, but preferred over stemming as lemmatization does morphological analysis of the word. Before lemmatizing, sentence should be tokenized into words.

```python
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize

tokenized_sentence = word_tokenize("NLP, a field blending linguistics, AI, and psychology, revolutionizes communication.")

lemmatizer = WordNetLemmatizer()
final_lemmatized_sentence = [lemmatizer.lemmatize(token) for token in tokenized_sentence]
final_stemmed_sentence = [stemmer.stem(token) for token in tokenized_sentence]
```

### Pointwise Mutual Information using NLTK

Co-occurence matrix for retrieving the number of times tokens appear together is a naive approach as it only conveys information about how many it has appeared. But, comparing how many times the token as appear alone in the entire text versus how many times it has appear as a part of bigram gives more useful insight. pmi is one or the useful technique to form bigrams in relative manner

```python
from nltk import BigramCollocationFinder
from nltk.collocations import BigramAssocMeasures
```

```python
list_of_stings = ['the cat sat', 'the cat sat in the hat', 'the cat with the hat']

# split the strings into tokens and form one single corpus
corpus = [x for sublist in list_of_stings for x in sublist.split()]
corpus
```

```python
bigram_measures = BigramAssocMeasures()

bigrams = BigramCollocationFinder.from_words(corpus)
```

```python
# score them
scored = bigrams.score_ngrams(bigram_measures.pmi)
scored
```

# WordCloud using NLTK

A data visualization technique, a graphical representation of text data where words from a text document are displayed in varying sizes, with the most frequently occurring words appearing larger.

The Brown Corpus was the first million-word electronic corpus of English, created in 1961 at Brown University. This corpus contains text from 500 sources, and the sources have been categorized by genre, such as news, editorial, and so on.

```python
from nltk.corpus import brown
nltk.download('brown')
brown.categories()
```

```python
brown_text = brown.words(categories='news')

print(len(brown_text))
brown_text
```

```python
brown_sentences = ' '.join(brown_text)
brown_sentences
```

Generating WordCloud

```python
wordcloud = WordCloud(background_color="white", max_font_size=100).generate(brown_sentences)
plt.figure(figsize=(10,8))

plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from wordcloud import STOPWORDS

stopwords = set(STOPWORDS)
len(stopwords)
```

wordcloud after removing stopwords

```python
wordcloud=WordCloud(stopwords=stopwords, background_color='white', max_font_size=100).generate(brown_sentences)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

Limiting number of words for the wordcloud

```python
wordcloud=WordCloud(stopwords=stopwords, background_color='white', max_font_size=100, max_words=100).generate(brown_sentences)
plt.figure(figsize=(10,8))
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis("off")
plt.show()
```

# spaCy

Free open source library for NLP. It includes NER (Named Entity Recognition) , POS tagging, etc.

"en_core_web_sm" is a small language model provided by spaCy library, its a pretrained model that comes with spaCy and can be used for various NLP tasks.

```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
text = "Machine learning is a field of artificial intelligence that empowers computers to learn from data"

# the doc object is a type of container that contains all the linguistic information like tokens, annotations, and relationships about the text.
doc = nlp(text)

for token in doc:
  print(token.text)
```

Stopwords in spaCy

```python
stopwords = spacy.lang.en.stop_words.STOP_WORDS

print("No of stopwords in SpaCy: ",len(stopwords))
```

```python
# printing 10 stopwords

for s in list(stopwords)[1:10]:
    print(s)
```

printing tokens which are not stop word

```python
[token for token in doc if not token.is_stop]
```

Finding similarity ( pointwise mutual info) between two texts using spaCy:

```python
import spacy
nlp1 = spacy.load("en_core_web_sm")

doc1 = nlp1('assisted living communities near me')
doc2 = nlp1('list of assisted living facilities')
doc3 = nlp1('free puppy and kitty adoption')

print(doc1.similarity(doc2))
print(doc1.similarity(doc3))
```

NER and POS tagging using spaCY

```python
nlp2 = spacy.load("en_core_web_sm")

text = "NLP, a field blending linguistics, AI, and psychology, revolutionizes communication."
doc = nlp2(text)
```

POS tagging:

```python
print("\nPOS Tags:")
for token in doc:
    print(f"{token.text}: {token.pos_}")
```

Finding entities:

```python
print("\nNamed Entities:")
for ent in doc.ents:
    print(f"{ent.text}: {ent.label_}")
```
